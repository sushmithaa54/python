```python
# 18)Numbers divisible by 3 from 1 to 20
n = 1

while n <= 20:
    if n % 3 == 0:
        print(n, end=' ')
    n = n + 1
```

```python
# 19) Numbers divisible by 3 and 5 from 1 to 20
n = 1

while n <= 20:
    if n % 3 == 0 and n % 5 == 0:
        print(n, end=' ')
    n = n + 1
```

```python
# 20) Numbers whose last digit is between 6 and 9 (25-29)
n1 = 25

while n1 <= 29:
    n2 = n1 % 10
    if n2 >= 6 and n2 <= 9:
        print(n1)
    n1 = n1 + 1
```

```python
# 21) Print even digits from number
n = 348756

while n > 0:
    digit = n % 10
    if digit % 2 == 0:
        print(digit, end=' ')
    n = n // 10
```

```python
# 22) Print odd digits from number
n = 348756

while n > 0:
    digit = n % 10
    if digit % 2 != 0:
        print(digit, end=' ')
    n = n // 10
```

```python
# 23) Print even digits greater than 4
n = 348756

while n > 0:
    digit = n % 10
    if digit % 2 == 0 and digit > 4:
        print(digit, end=' ')
    n = n // 10
```

```python
# 24) Sum of digits of a number
n = 5687
total = 0

while n > 0:
    digit = n % 10
    total = total + digit
    n = n // 10

print(total)
```

```python
# 25) Sum of even digits
n = 5687
total = 0

while n > 0:
    digit = n % 10
    if digit % 2 == 0:
        total = total + digit
    n = n // 10

print(total)
```

```python
# 26) Product of odd digits and sum of even digits
n = 5872
pro = 1
sum = 0

while n > 0:
    digit = n % 10
    if digit % 2 == 0:
        sum = sum + digit
    else:
        pro = pro * digit
    n = n // 10

print(sum)
print(pro)
```

```python
# 27) Sum and product of all digits
n = 5872
product = 1
sum = 0

while n > 0:
    digit = n % 10
    product = product * digit
    sum = sum + digit
    n = n // 10

print(f"product of digits : {product}")
print(f"sum of digits : {sum}")
```
