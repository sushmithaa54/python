```python
# example for 'w' and 'a' mode without using with statement
import csv
file=open('python.csv','w',newline='')
writer=csv.writer(file)
writer.writerow(['ename','sal','deptno'])
writer.writerow(['scott',2333,10])
file.close()
```

```python
# example for 'w' and 'a' mode using with statement
import csv
with open('python.csv','w',newline='') as file:
    writer=csv.writer(file)
    writer.writerow(['ename','sal','deptno'])
    writer.writerow(['scott',2333,20])
```

```python

# example for 'r' mode without using with statement
import csv
file=open('python.csv','r')
reader=csv.reader(file)
for row in reader:
    print(row)
file.close()
```

```python

# example for 'r' mode using with statement
import csv
with open('python.csv','r') as file:
    reader=csv.reader(file)
    for row in reader:
        print(row)
```

```python
# example for 'w+'mode without using with statement
import csv
file=open('python.csv','w+',newline='')
writer=csv.writer(file)
writer.writerow(['usn','name','cgpa'])
writer.writerow([120,'ram',10])
file.seek(0)
reader=csv.reader(file)
for row in reader:
    print(row)
file.close()
```

```python
# example for 'w+' mode using with statement
import csv
with open('python.csv','w+',newline='') as file:
    writer=csv.writer(file)
    writer.writerow(['usn','name','cgpa'])
    writer.writerow([120,'ram',10])
    file.seek(0)
    reader=csv.reader(file)
    for row in reader:
        print(row)

```

```python
# example for 'r+' mode without using with statement
import csv
file=open('python.csv','r+')
reader=csv.reader(file)
for row in reader:
    print(row)
writer=csv.writer(file)
writer.writerow(['scott',2333,20])
file.close()
```
