```python
# WAP to perform ckech balance,deposite and withdraw operations
class bank:
    banckname='icici'
    
    def details(self,cname,balance):
        self.cname=cname
        self.balance=balance
    def deposite(self,amount):
        if amount>0:
            self.balance=self.balance+amount
            print(f'{amount} has been deposited')
            print(f'after deposite uploade balance :{self.balance}')
        else:
            print('amount should be greater than zero')
    def withdraw(self,amount):
        if self.balance>amount:
            self.balance=self.balance-amount
            print(f'{amount} has been debited')
            print(f'after withdraw updated balance is:{self.balance}')
        else:
            print('insuficient balance')
    def display(self):
        print(f'customer name:{self.cname}')
        print(f'initial bank balance is:{self.balance}')
c1=bank()
c1.details('smith',98000)
c1.display()
c1.deposite(1000)
c1.withdraw(2000)
print()
c2=bank()
c2.details('king',450000)
c2.display()
c2.deposite(2000)
c2.withdraw(50000)
    
```
