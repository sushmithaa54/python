```python
ch=input()
if 'A'<= ch <= 'B':
    print('upper')
elif 'a'<=ch<='b':
    print('lower')
```

```python
n=int(input())
if n%3==0 and n%5==0:
    print('divisible by both')
elif n%3==0:
    print('dividible by 3')
elif n%5==0:
    print('divisible by 5')
```

```python
ch=input()
if 'a'<=ch<='b' or 'A'<=ch<='B':
    print('alphabet')
elif '0'<=ch<='9':
    print('digit')
    
```

```python
ch=input()
if 'a'<=ch<='z':
    print('upper')
elif 'A'<=ch<='Z':
    print('lower')
elif '0'<=ch<='9':
    print('digit')
else:
    print('symbol')
```
