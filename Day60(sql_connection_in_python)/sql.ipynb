```python
# create a table
import sqlite3
conn=sqlite3.connect('ram_ram.db')
cursor=conn.cursor()
cursor.execute('''
create table employee (empno intege,ename text,sal real)
               ''')
conn.commit()
conn.close()
```

```python
# insert data to the table
import sqlite3
conn=sqlite3.connect('ram_ram.db')
cursor=conn.cursor()
cursor.execute(''' insert into employee (empno,ename,sal) values(120,'king',1000)''')
cursor.execute(''' insert into employee (empno,ename,sal) values(121,'scott',7000)''')
conn.commit()
conn.close()
```

```python
# read from the table
import sqlite3
conn=sqlite3.connect('ram_ram.db')
cursor=conn.cursor()
cursor.execute('select * from employee')
rows=cursor.fetchall()
for row in rows:
    print(row)
conn.commit()
conn.close()
```

```python
# update the values

import sqlite3
conn=sqlite3.connect('ram_ram.db')
cursor=conn.cursor()
cursor.execute(''' update employee set sal=98000 where ename='king' ''')
conn.commit()
conn.close()
```

```python
# delete the values
import sqlite3
conn=sqlite3.connect('ram_ram.db')
cursor=conn.cursor()
cursor.execute(''' delete from employee where ename='scott' ''')
conn.commit()
conn.close()
```
