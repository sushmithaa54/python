```python
n=int(input())
if n>5:
    if n%2==0:
        print("even")
    else:
        print("odd")
else:
    print(n)
```

```python
ch=input()
if 'A'<=ch<='Z' or 'a'<=ch<='z':
    if ch in 'aeiouAEIOU':
        print("vowel")
    else:
        print("consonent")
else:
    if '0'<=ch<='9':
        print("digit")
    else:
        print("not a digit")
```
