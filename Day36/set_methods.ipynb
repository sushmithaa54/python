```python
# 1) ADD METHOD
s1={'s','m'}
s1.add('h')
print(s1)
```

```python
# 2) REMOVE METHOD
s1={'s','m'}
s1.remove('s')
print(s1)
```

```python
# 3) UPDATE METHOD
set1={10,20,30,40}
set2={30,40,50,60}
set1.update(set2)
print(set1)
```

```python
# 4) POP METHOD
s1={10,20,30,40,89}
s1.pop()
print(s1)
```

```python
# 5) UNION METHOD
s1={10,20,30,40,89}
s2={70,60,30}
res=s1.union(s2)
print(res)
```

```python
# 6) INTERSECTION METHOD
s1={10,20,30,40,89}
s2={70,60,30}
res=s1.intersection(s2)
print(res)
```

```python
# 7) DIFFERENCE METHOD
s1={10,20,30,40,89}
s2={70,60,30}
res=s1.difference(s2)
print(res)
```

```python
# 8) SYMMETRIC_DIFFERENCE METHOD
s1={10,20,30,40,89}
s2={70,60,30}
res=s1.symmetric_difference(s2)
print(res)
```
