```python
# 38 sum of all numbers present in the list
l=[10,45,68,95,40]
i=0
result=0
while i<len(l):
    result=result+l[i]
    i=i+1
print(f'sum of list numbers is:{result}')
```

```python
# 39 sum of even digits present in a list
l=[10,45,68,95,40]
i=0
result=0
while i<len(l):
    if l[i]%2==0:
       result=result+l[i]
    i=i+1
print(f'sum of even digits {result}')
```

```python
# 42 find factorial of a number
num=5
fact=1
while num>0:
    fact=fact*num
    num=num-1
print(fact)
```
