```python
# print cube of number
def cube(num):
    return num**3
res=cube(3)
print(res)
```

```python
# to check given number is even or odd
def number(n):
    if n%2==0:
        return f'{n} is even'
    else:
        return f'{n} is odd'
res=(number(900))
print(res)
```
