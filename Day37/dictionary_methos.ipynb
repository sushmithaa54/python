```python
# 1) GET METHOD
data={'ename':'smith','salary':89000}
res=data.get('ename')
print(res)

data={'ename':'smith','salary':89000}
res=data.get('job')
print(res)


```

```python
# 2) UPDATE METHOD
data={'ename':'smith','salary':89000}
data.update({'ename':'martin'})
print(data)

data={'ename':'smith','salary':89000}
data.update({'job':'analyst'})
print(data)

data={'ename':'smith','salary':89000}
data.update(salary=4000)
print(data)



```

```python
# 3) ITEMS METHOD
data={'ename':'smith','salary':89000}
print(data.items())
```

```python
# 4) KEYS METHOD
data={'ename':'smith','salary':89000}
print(data.keys())
```

```python
# 5) VALUES METHOD
data={'ename':'smith','salary':89000}
print(data.values())
```

```python
# 6) POP METHOD
data={'ename':'smith','salary':89000}
data.pop('ename')
print(data)
```

```python
# 7) POPITEM METHOD
data={'ename':'smith','salary':89000}
data.popitem()
print(data)
```

```python
# 8) CLEAR METHOD
data={'ename':'smith','salary':89000}
data.clear()
print(data)
```
