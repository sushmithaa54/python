```python
# OBJECT METHOD

class bankaccount:
    def __init__(self,name,balance):
        self.name=name
        self.balance=balance
    def deposit(self,amount):
        print(f'name of the account holder:{self.name}')
        print(f'balance before deposit:{self.balance}')
        if amount>0:
            self.balance=self.balance+amount
            print(f'balance after deposit:{self.balance}')
        else:
            print('amount should be more than 0')
b1=bankaccount('ranjan',60000)
b1.deposit(1000)
        
```

```python
# CLASS METHOD

class bankaccount:
    b_name='icici'
    def __init__(self,name,balance):
        self.name=name
        self.balance=balance
    
    
    @classmethod
    def change_name(cls,new_name):
        print(f'bank name before change:{cls.b_name}')
        cls.b_name=new_name
        print(f'bank name after change:{cls.b_name}')
    
    def deposit(self,amount):
        print(f'name of the account holder:{self.name}')
        print(f'balance before deposit:{self.balance}')
        if amount>0:
            self.balance=self.balance+amount
            print(f'balance after deposit:{self.balance}')
        else:
            print('amount should be more than 0')
    

b1=bankaccount('chethan',4000)
b1.change_name('canara')
b1.deposit(200)
    
```

```python
# WAP TO CREATE INSTANCE METHOD,CLASS METHOD AND STATIC METHOD TO DISPLAY STUDENT NAME,MARKS AND SCHOOL NAME 
# 1.INSTANCE VARIABLES(STUDENT NAME AND MARKS)
# 2.CLASS VARIABLES(SCHOOL NAME)
# 3.INSTANCE METHOD(ADD BONUS MARKS)
# 4.CLASS METHOD(CHANGE SCHOOL NAME)
# 5.STATIC METHOD(CHECK PASS OR FAIL :- PASS IF MARKS IS GREATER THAN OR EQUAL TO 35(MARKS>=35))

class student:
    school_name='jspiders'
    def __init__(self,sname,marks):
        self.sname=sname
        self.marks=marks
    
    def add_bonus(self,bonus):
        print(f'marks before adding bonus marks:{self.marks}')
        self.marks=self.marks+bonus
        print(f'marks after adding bonus marks:{self.marks}')
        
    @classmethod
    def change_name(cls,new_name):
        print(f'school name before change:{cls.school_name}')
        cls.school_name=new_name
        print(f'school name before change:{cls.school_name}')
        
        
    @staticmethod
    def check(marks):
        return marks>=35
    
    def display(self):
        print(f'student name:{self.sname},marks:{self.marks} and school name:{self.school_name}')
        if student.check(self.marks):
            print('RESULT:PASS')
        else:
            print('RESULT:FAIL')

s=student('sagar',20)
s.display()
s.add_bonus(10)
s.change_name('pyspiders')
        
```
