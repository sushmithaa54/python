```python
# 17 check no. is palindrome or not
n=121
temp=n
rev=0
for i in str(n):
    if temp>0:
        digit=temp%10
        rev=rev*10+digit
        temp=temp//10
if rev==n:
    print(' palindrome ')
else:
    print(' not palindrome ')
```

```python
# 18 given no is armstrong no or not
n=153
temp=n
sum=0
length=len(str(n))
for i in str(n):
    if temp>0:
        digit=temp%10
        sum=sum+digit**length
    temp=temp//10
if n==sum:
    print('amstrong')
else:
    print('not amstrong')
```
