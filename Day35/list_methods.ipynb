```python
# 1) APPEND METHOD

list=[10,20,30,40]
list.append(50)
print(list)

list=[10,20,30,40]
list.append('hello')
print(list)
```

```python
# 2) EXTEND METHOD

list=[10,20,30,40]
list.extend([50,60,70])
print(list)

list=[10,20,30,40]
list.extend(['hello','hiii'])
print(list)
```

```python
# 3) COUNT METHOD

list=[10,20,30,20,20,80,40]
res=list.count(20)
print(res)
```

```python
# 4) INDEX METHOD

list=[10,20,30,40,50]
res=list.index(50)
print(res)

# example:-

# wap to find position of second occurance of specified element
list=[10,20,50,30,40,50,50]
first=list.index(50)
second=list.index(50,first+1)
print(second)

```

```python
# 5) REVERSE METHOD

list=[10,20,30,20,20,80,40]
list.reverse()
print(list)
```

```python
# 6) SORT METHOD
# for acsending
list=[10,20,30,20,20,80,40]
list.sort()
print(list)

#for decsending
list=[10,20,30,20,20,80,40]
list.sort(reverse=True)
print(list)

# for acsending
list=['sql','python','java']
list.sort()
print(list)

# for descending
list=['sql','python','java']
list.sort(reverse=True)
print(list)

# bassed on length highest to lowest
list=['sql','python','java']
list.sort(key=len)
print(list)

# bassed on length lowest to highest
list=['sql','python','java']
list.sort(key=len,reverse=True)
print(list)


```

```python
# 7) POP METHOD

list=['sql','python','java']
list.sort(reverse=True)
print(list)


list=[10,20,30,20,20,80,40]
list.pop(4)
print(list)
```

```python
# 8) REMOVE METHOD

list=[10,20,30,20,20,80,40]
list.remove(20)
print(list)
```

```python
# 9) CLEAR METHOD

list=[10,20,30,20,20,80,40]
list.clear()
print(list)
```
