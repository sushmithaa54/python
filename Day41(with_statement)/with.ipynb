```python
# write mode using wuth statement
with open('python.txt','w') as file:
    file.write('hello python')
```

```python
# read mode using wuth statement
with open('python.txt','r') as file:
    display=file.read()
    print(display)
```

```python
# read line (first line only)
with open('python.txt','r') as file:
    display=file.readline()
    print(display)
```

```python
# read lines (all lines in the form of list)
with open('python.txt','r') as file:
    display=file.readlines()
    print(display)
```

```python
# read lines (perticular line by using index))
with open('python.txt','r') as file:
    display=file.readlines()[2]
    print(display)
```

```python
# append mode using wuth statement
with open('python.txt','a') as file:
    file.write('\nwe are good couple....')
```

```python
# w+ mode using wuth statement
with open('python.txt','w+') as file:
    file.write('\n please nazar mat lagana')
    file.seek(0)
    display=file.read()
    print(display)
    
```

```python
# r+ mode using wuth statement
with open('python.txt','r+') as file:
    display=file.read()
    print(f'line before add:{display}')
    file.write('\n becouse nazar is real.....')
    file.seek(0)
    display1=file.read()
    print(f'line after add:{display1}')
    
```

```python
# a+ mode using wuth statement
with open('python.txt','a+') as file:
    file.write('\n yahhhhhhh we are best couple')
    file.seek(0)
    display=file.read()
    print(display)
    
```
