```python
num=5
i=1
count=0
while i<=num:
    if num%i==0:
        count=count+1
    i+=1
if count==2:
    print(f'{num} is a prime number')
else:
    print(f'{num} is not a prime number')
```

```python
num=153
temp=num
sum_cubes=0
length=len(str('num'))
while num>0:
    digit=num%10
    sum_cubes=sum_cubes+digit**length
    num=num//10
if temp==sum_cubes:
    print(f'{temp} is an armstrong numberd')
else:
    print(f'{temp} is not an armstrong numberd')
```
