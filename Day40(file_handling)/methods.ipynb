```python
# 1) write method
file=open('python1.txt','w')
content=file.write('writting a data')
file.close()
```

```python
# 2) read method
file=open('python1.txt','r')
display=file.read()
print(display)
file.close()
```

```python
# 3) append method
file=open('python1.txt','a')
content=file.write('\nthis is the first line')
file.close()
```

```python
# to display the position of the cursor we use "tell" method
file=open('python1.txt','r')
display=file.read()
print(display)
position=file.tell()
print(f'\nposition of cursor is : {position}')
file.close()
```
