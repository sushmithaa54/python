```python
# 31 find higthest digit in the given number
n=12345
highest=0
while n>0:
    digit=n%10
    if digit>highest:
        highest=digit
    n=n//10
print(f'{highest} is higthest number')
```

```python
# 32 find lowest digit in the given number
n=12345
lowest=9
while n>0:
    digit=n%10
    if digit<lowest:
        lowest=digit
    n=n//10
print(f'{lowest} is lowest number')
```

```python
## 33 find sum of the  digit in the given number
n=1245
sum_digit=0
while n>0:
    digit=n%10
    sum_digit=sum_digit+digit
    n=n//10
if sum_digit%2==0:
    print(f'{sum_digit} is even number')
else:
    print(f'{sum_digit} is odd number')
```
