```python
# wap to divide 2 numbers and handle zero division error
try:
    a=40
    b=0
    res=a/b
    print(res)
except ZeroDivisionError:
    print(f'{a} is not divisible by zero')
```

```python
# wap to acces ist element,if there is index error display index value out of range.
try:
    list=[10,20,30]
    print(list[4])
except IndexError:
    print('index value out of range')
```
