```python
class employee:
    def __init__(self,ename,job):
        self.ename=ename
        self.job=job
    def display(self):
        print(f'employee name:{self.ename} and job:{self.job}')
d1=employee('nitya','fassion designer')
d1.display()
d2=employee('manthan','photographer')
d2.display()
```
