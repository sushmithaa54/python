```python
class employee:
    def __init__(self,ename,job,monthly_sal):
        self.ename=ename
        self.job=job
        self.monthly_sal=monthly_sal
    def anual_sal(self):
        annual_sal=self.monthly_sal*12
        print(f'employee name:{self.ename} , job:{self.job}, monthly salary{self.monthly_sal} and annual salary {annual_sal} ')
d1=employee('nitya','fassion designer',98000)
d1.anual_sal()
d2=employee('manthan','photographer',45000)
d2.anual_sal()
```
