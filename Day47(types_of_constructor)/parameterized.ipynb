```python
class employee:
    def __init__(self,ename,job):
        self.ename=ename
        self.job=job
    def display(self):
        print(f'employee name{self.ename} and job:{self.job}')

e1=employee('sagar','analyst')
e1.display()
print()
e2=employee('king','salesman')
e2.display()
```

```python
class bank:
    def __init__(self,cname,id):
        self.cname=cname
        self.id=id
    def display(self):
        print(f'account holder name is {self.cname} and refference id is {self.id} ')
c1=bank('chaaya',101)
c1.display()
```
