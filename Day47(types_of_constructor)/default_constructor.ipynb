```python
#WAP to create default constructor and to diaplay employeee name and salary
class employee:
    def read(self,ename='sagar',sal=2000):
        self.ename=ename
        self.sal=sal
emp1=employee()
emp1.read()
print(f'employee name{emp1.ename}and salary {emp1.sal}')

emp2=employee()
emp2.read('sumanth',8000)
print(f'employee name{emp2.ename}and salary {emp2.sal}')

```
