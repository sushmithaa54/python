```python


class bank:
    def __init__(self,name,balance):
        self.name=name
        self.balance=balance
        
    def display(self):
        print(f'customer name:{self.name}\nbalance:{self.balance}')
        
obj=bank('ram',20000)
obj.display()
print()
print(f'accessing outside class name:{obj.name}')
print(f'accessing outside class balance:{obj.balance}')

obj.name='krishna'
obj.balance=98000
print()
print(f'name after modification:{obj.name}')
print(f'balance after modification:{obj.balance}')
```
