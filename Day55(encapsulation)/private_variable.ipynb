```python
class bank:
    def __init__(self,name,balance):
        self.__name=name
        self.__balance=balance
        
    def _showdetails(self):
        print(f'customer name:{self.__name}\nbalance:{self.__balance}')
        
class customer(bank):
    def display(self):
       self._showdetails()
        
obj=customer('ram',20000)
obj.display()


# print(obj.__balance)
# using name mangling
print(obj._bank__balance)
```
