```python
class employee:
    def __init__(self,empno,ename,salary):
        self.empno=empno
        self.ename=ename
        self.__salary=salary
    # getter
    def get_salary(self):
        return self.__salary
    # setter
    def set_salary(self,new):
        if new>0:
            self.__salary=new
        else:
            print('salary more than zero')
            
e=employee('12','ram',234567)
print(f'emplyee number:{e.empno}\nname:{e.ename}\nsalary:{e.get_salary()}')

e.set_salary(98000)
print(f'updated salary:{e.get_salary()}')
        
```

```python
class bankaccount:
    def __init__(self,balance):
        self.__balance=balance
        
    @property
    def balance(self):
        return self.__balance
    
    @balance.setter
    def balance(self,new):
        if isinstance(new,int):
            self.__balance=new
        else:
            raise ValueError('invalid formate data')
        
c1=bankaccount(30000)
c1.balance=45000
print(c1.balance)
```
