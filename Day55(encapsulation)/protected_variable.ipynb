```python
class bank:
    def __init__(self,name,balance):
        self._name=name
        self._balance=balance
        
    def display(self):
        print(f'customer name:{self._name}\nbalance:{self._balance}')
        
obj=bank('ram',20000)
obj.display()
print()
print(f'accessing outside class name:{obj._name}')
print(f'accessing outside class balance:{obj._balance}')
```
