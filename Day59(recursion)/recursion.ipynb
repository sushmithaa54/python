```python
# 1
def factorial(num):
    if num==0:
        return 1
    else:
        return num*factorial(num-1)
print(factorial(5))
```

```python
# 2
def total(num):
    if num==0:
        return 0
    else:
        return num+total(num-1)
print(total(10))
```

```python
# 3
def count_num(num):
    if num==0:
        return 0
    else:
        return 1+count_num(num//10)
print(count_num(8976))
```

```python
# 4
def sum_digits(num):
    if num==0:
        return 0
    else:
        return num%10 + sum_digits(num//10)
print(sum_digits(5678))
```

```python
# 5
def palindrome(num,rev=0):
    if num==0:
        return rev
    else:
        return(palindrome(num//10,rev*10+num%10))
num=121
print(num==palindrome(num))
```
