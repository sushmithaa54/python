```python
# 1 wap to print no.s from 1 to 5 except 3
for i in range(1,6):
    if i==3:
        continue
    print(i,end=" ")
```

```python
# 2 wap to print only odd no.s from 1 to 10
for i in range(1,11):
    if i%2==0:
        continue
    print(i,end=" ")
```

```python
# 3 wap to print only even no.s from 1 to 10
for i in range(1,11):
    if i%2!=0:
        continue
    print(i,end=" ")
```
