```python
# to reverse a number

def reverse_num(num):
    rev=0
    for i in str(num):
        digit=num%10
        rev=rev*10+digit
        num=num//10
    return rev
print(reverse_num(2345))
        
        
```

```python
# to find sum of digits
def sum_digits(num):
    sum=0
    for i in str(num):
        sum=sum+int(i)
    return sum
print(sum_digits(1235))
```

```python

# to check palindrom or not
def palindrom(num):
    temp=num
    rev=0
    for i in str(temp):
        digit=temp%10
        rev=rev*10+digit
        temp=temp//10
    if rev==num:
        return f'palindrome'
    else:
        return f'not'
        
print(palindrom(1221))


```

```python
# larger of two numbers
def large(num1,num2):
    if num1>num2:
        return f'{num1} is larger'
    elif num2>num1:
        return f'{num2} is larger'
print(large(2,3))
    
```

```python
# armstrong number
def arm(num):
    length=len(str(num))
    pro=0
    for i in str(num):
        pro=pro+int(i)**length
    if pro==num:
        return f'{num} is armstrong number'
print(arm(153))
```

```python
# strong number or not
def strong(num):
    sum_fact=0
    for i in str(num):
        fact=1
        for x in range(1,int(i)+1):
            fact=fact*x
        sum_fact=sum_fact+fact
    if sum_fact==num:
        return f'{num} is strong number'
print(strong(145)) 
```

```python
# perfect number or not
def perfect(num):
    sum=0
    for i in range(1,num):
        if num%i==0:
            sum=sum+i
    if sum==num:
        return f'{num} is perfect number'
print(perfect(6))
```

```python
# count the digits
def digits(n):
    count=0
    for i in str(n):
        count=count+1
    return count
print(digits(1234563456))
```

```python
# sum of n natural numbers (5)
def sum_natu(n):
    sum=0
    for i in range(n+1):
        sum=sum+i
    return sum
print(sum_natu(5))
    
```

```python
# spy number or not
def spy(n):
    sum=0
    pro=1
    for i in str(n):
        sum=sum+int(i)
        pro=pro*int(i)
    if sum==pro:
        return f'{n} is spy number'
print(spy(128))
    
```

```python
# prime number or not
def prime(num):
    count=0
    for i in range(1,num+1):
        if num%i==0:
            count=count+1
    if count==2:
        return f'{num} is prime number'
print(prime(9))
```

```python
# prime
def prime(num):
    count=0
    i=1
    while i<=num:
        if num%i==0:
            count=count+1
        i=i+1
    if count==2:
        return f'{num} is prime number'
print(prime(5))
        
```
