```python
# 1) FIND METHOD
msg='python'
print(msg.find('t'))
print(msg.find('n'))
print(msg.find('j'))
```

```python
# 2) RINDEX METHOD
msg='python python python'
print(msg.rindex('h'))
```

```python
# 3) REPLACE METHOD
msg='welcome to python'
print(msg.replace('o','*'))
print(msg.replace('o','*',1))
print(msg.replace('o','*',2))
print(msg.replace('o','*',10))
```

```python
# 4) starts with method
msg='python'
print(msg.startswith('p'))
print(msg.startswith('P'))
print(msg.startswith('j'))
```

```python
# 4) ends with method
msg='python'
print(msg.endswith('n'))
print(msg.endswith('N'))
print(msg.endswith('j'))
```

```python
# 5) SPLIT METHOD
msg='welcome to python class'
print(msg.split())
print(msg.split('o'))


# examples
# 1) wap to display extension of file
msg='python.txt'
extension=msg.split('.')
print(extension[-1])


# 2) wap to display name of file
msg='python.txt'
extension=msg.split('.')
print(extension[0])

# 3)  wap to display domain of given url
msg='https://pyspiders.com'
extension=msg.split('://')
print(extension[-1])

# 4)  wap to display protocal of given url
msg='https://pyspiders.com'
extension=msg.split('://')
print(extension[0])
```

```python
# 6) JOIN METHOD

words=['HELLO','PYTHON']
result='*'.join(words)
print(result)
```

```python
# 7) ISALPHA METHOD

msg='python'
print(msg.isalpha())

msg='python113'
print(msg.isalpha())

```

```python
# 8) ISALNUM METHOD

msg='python123'
print(msg.isalnum())

msg='python'
print(msg.isalnum())

msg='python***123'
print(msg.isalnum())



```

```python
# 9) ISLOWER METHOD

msg='python'
print(msg.islower())

msg='Python'
print(msg.islower())

```

```python
# 10) ISUPPER METHOD

msg='PYTHON'
print(msg.isupper())

msg='Python'
print(msg.isupper())

```

```python
# 11) ISDIGIT METHOD

msg='1234567890'
print(msg.isdigit())

msg='Python123456789'
print(msg.isdigit())

```
