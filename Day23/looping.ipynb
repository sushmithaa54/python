```python
# 28) to reverse the given number
num=1234
rev=0
while num>0:
    digit=num%10
    rev=rev*10+digit
    num=num//10
print(rev)
```

```python
# 29) to check whether the given number is palindrom or not
num=int(input('enter a number:'))
rev=0
temp=num
while temp>0:
    digit=temp%10
    rev=rev*10+digit
    temp=temp//10
if num==rev:
    print(f'{num} is a palindrom number')
else:
    print(f'{num} is not a palindrom number')
```
