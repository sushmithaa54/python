```python
# method overloading
# # 1
class demo:
   def add(self,*args):
       return sum(args)
obj=demo()
print(obj.add(10,20,4))
print(obj.add(10,20,4,8,90,98))
print(obj.add(10,20,44,33,555,8000))
```

```python
# 2
class demo:
    def add(self,*args):
        res=0
        for i in args:
            res=res+i
        return res
    
obj=demo()
print(obj.add(12,45,6,7,3,4,90))
print(obj.add(12,45,3,4,90))
print(obj.add(12,45,6,7,3,4,90,2346))
```

```python
# 3
class demo:
    def multiply(self,*args):
        res=1
        for i in args:
            res*=i
        return res

obj=demo()
print(obj.multiply(10,20))
print(obj.multiply(10,20,200))
print(obj.multiply(10,20,200,40000))
```

```python
# methos overriding

# 1
class parent:
    def display(self):
        print('this is parent class')
class child(parent):
    def display(self):
        print('this is child class')
        
obj=child()
obj.display()
```

```python
# 2
class parent:
    def display(self):
        print('this is parent class')
class child(parent):
    def display(self):
        super().display()
        print('this is child class')
        
obj=child()
obj.display()
```

```python
# 3

class bank_acct:
    def __init__(self,name,balance=0):
        self.name=name
        self.balance=balance
    def deposit(self,amount):
        self.balance+=amount
        print(f'{amount} hs been deposited and updated balance : {self.balance}')
    
    def withdraw(self,amount):
        if amount>self.balance:
            print('insufficint bank balance')
        elif amount<self.balance:
            self.balance-=amount
            print(f'{amount} has been deposited and updated amount : {self.balance}')
            
    def check_bal(self):
        print(f'initial bank balance : {self.balance}')
        
class saving_acct(bank_acct):
    def withdraw(self, amount):
        if amount>10000:
            print('limit exceeded ! max withdrawal is 1000.')
        else:
            super().withdraw(amount)

acc1=bank_acct('jeeva',500000)
acc2=saving_acct('sush',300000)

acc1.check_bal()
acc1.deposit(200)
acc1.withdraw(900)
print()
acc2.check_bal()
acc2.deposit(200)
acc2.withdraw(900)
acc2.withdraw(299999)

    
```
